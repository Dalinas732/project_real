```python
"""
This moduel is a proof of concept how taking my zeros that obviously should not be there, and correcting them to the actual value we would expect. Only good for a single year.
"""
from DataCollection import Temperature_Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("cleandata.xlsx", sheet_name=None)
tp = Temperature_Data(data)
```

```python
years = tp.years
year = "1978"
df = tp.dfs[year].astype(float)
```

```python
### This checks to see if there is more than two zeros in a given month. If so this is most likely a data problem and not actual zeros.
def check_zeros(data):
    found = False
    count = 0
    for i in range (data.size-1):
        if data[i] == 0:
            count+=1
    if count >=2:
        return True
    else:
        return False
```

```python
# Define the Excel filename
file_name = "no_zeros_data.xlsx"
```

```python
year = "1978"
months = np.arange(12) + 1
for month in months:
    lows = tp.collect_month(month,year,"low")
    avgs = tp.collect_month(month,year,"average")
    highs = tp.collect_month(month,year,"high")
    tp.check_length_of_month(lows.size,month,year)
    tp.check_length_of_month(avgs.size,month,year)
    tp.check_length_of_month(highs.size,month,year)
    if check_zeros(lows): ### If zeros are present change the value based on the high and average
        for i in range(lows.size):
            if lows[i] == 0:
                newlow = round(2*avgs[i] - highs[i],3)
                lows[i] = newlow
                assert lows[i] == newlow, f"The low was not changed for month = {month} and index = {i}."     
    pd_lows = np.append(lows,[float(np.nan)]*(31-lows.size))
    df.iloc[:,(3*month-1)] = pd_lows
with pd.ExcelWriter(file_name, engine='xlsxwriter') as writer: ### Testing a writing function to produce a new datasheet with non-zero lows.
    df.to_excel(writer, sheet_name=year, index=False)
```
