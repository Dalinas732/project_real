```python
"""
This moduel is a proof of concept using standard deviations to actually see what you would expect the temperature to be rather than just the average.
"""
from DataCollection import Temperature_Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("cleandata.xlsx", sheet_name=None)
t = Temperature_Data(data)
```

```python
### Finding all daily Standard Deviations for January highs
std = []
avg = []
month = 3
for i in range(31):
    day = i+1
    std.append(np.std(t.collect_day(month,day,0)))
    avg.append(np.mean(t.collect_day(month,day,0)))
stds = np.array(std)
avg_temp = np.array(avg)
avg_std = np.mean(stds)
print("+-",avg_std/2)
```

```python
x = np.arange(31) + 1

plt.plot(x,stds)
plt.grid()
plt.title(f"Standard Deviations of Daily Highs in month = {month} ")
plt.ylabel("Degrees F")
plt.xlabel("Day")
plt.show()
print("Standard Deviations of Standard Deviations...",np.mean(stds),"+-",np.std(stds))
```
