```python
### This reads the data as a dictionary of Pandas Data Frames.
dfs = pd.read_excel("rawdata.xlsx", sheet_name=None)  # Reads all sheets
print(dfs.keys())  # List of sheet names
```

```python
# Here I am relabeling the first column so it matches the rest
count = 0
years = []  ## I have made a list with all the year numbers as strings so I can easily call each dataframe with an index
for i in dfs.keys():
    df = dfs[i]
    df.rename(columns={"Max":"Max.0","Avg":"Avg.0","Min":"Min.0"}, inplace=True)
    count = count + 1
    years.append(i)
    
```

```python
### Figuring out how to recognize a nan
a = np.array(df.iloc[28:31,3])
print(np.isnan(a[0]))
```

```python
### Test run for calculating average high for 2023
i = 0
total = 0
N = 0
count_nan = 0
while 3*i < 36:
    high = np.array(df.iloc[:, 3*i]).astype(float)
    for temp in high:
        if np.isnan(temp) == False:
            total+=temp
            N+=1
        else:
            count_nan+=1
    i+=1
avg = total / N
print(avg,count_nan)
```

```python
### Let's store all the highs in an array and plot a histogram for February 20th
feb_20 = []
for year in years:
    df = dfs[year]
    high = np.array(df.iloc[19,3])
    feb_20.append(high)
feb_20 = np.array(feb_20)
print(feb_20)
```

```python
import matplotlib.pyplot as plt
# Plot the histogram
bins = np.linspace(np.min(feb_20),np.max(feb_20),feb_20.size)
plt.hist(feb_20, bins, edgecolor='black')

# Labels and title
plt.xlabel('Temperature')
plt.ylabel('# of Times')
plt.title('Highs for February 20th 1960-2024')

# Show the plot
plt.show()
```
